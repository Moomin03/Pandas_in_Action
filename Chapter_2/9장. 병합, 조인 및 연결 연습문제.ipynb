```python
import pandas as pd
```

```python
week1 = pd.read_csv(filepath_or_buffer='restaurant/week_1_sales.csv')
week2 = pd.read_csv(filepath_or_buffer='restaurant/week_2_sales.csv')
```

```python
customers = pd.read_csv(filepath_or_buffer='restaurant/customers.csv',
                        index_col='ID')
```

```python
food = pd.read_csv(filepath_or_buffer='restaurant/foods.csv',
                   index_col='Food ID')
```

```python
# 1번
first = pd.concat(objs=[week1, week2],
          keys=['Week 1', 'Week 2'])
first
```

```python
# 2번
week1.merge(right=week2,
            how='inner',
            on='Customer ID').drop_duplicates(subset=['Customer ID'])
```

```python
# 3번
week1.merge(right=week2,
            how='inner',
            on=['Customer ID', 'Food ID']).drop_duplicates(subset=['Customer ID'])
```

```python
# 4번
fourth = week1.merge(right=week2,
            how='outer',
            on=['Customer ID'],
            indicator=True)
fourth[fourth['_merge']=='left_only'].drop_duplicates(subset=['Customer ID'])
```

```python
fourth[fourth['_merge']=='right_only'].drop_duplicates(subset=['Customer ID'])
```
