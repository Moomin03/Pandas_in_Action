```python
# 먼저 자동차 데이터
cars_data = pd.read_csv('used_cars.csv')
cars_data.head()
```

```python
# 1번
cars_data.pivot_table(index='Fuel',
                      values='Price',
                      aggfunc='sum')
```

```python
# 2번
cars_data.pivot_table(index='Manufacturer',
                      values='Price',
                      columns='Transmission',
                      aggfunc='count',
                      margins=True)
```

```python
# 3번
cars_data.pivot_table(index=['Year', 'Fuel'],
                      columns=['Transmission'],
                      values='Price',
                      aggfunc='mean')
```

```python
# 4번
problem = cars_data.pivot_table(index=['Year', 'Fuel'],
                      columns=['Transmission'],
                      values='Price',
                      aggfunc='mean')
problem.stack()
```
