```python
# 1번
netflix = pd.read_csv(filepath_or_buffer='netflix.csv',
                      parse_dates=['date_added'])
netflix.head()
```

```python
netflix.info()
```

```python
netflix.nunique()
```

```python
netflix['type'] = netflix['type'].astype('category')
```

```python
# 2번
netflix[netflix['title']=='Limitless']
```

```python
# 3번
netflix[(netflix['director']=='Robert Rodriguez')&(netflix['type']=='Movie')]
```

```python
# 4번
netflix[(netflix['date_added']=='2019-07-31')|(netflix['director']=='Robert Altman')]
```

```python
# 5번
netflix[(netflix['director']=='Orson Welles')|(netflix['director']=='Aditya Kripalani')|(netflix['director']=='Sam Raimi')]
```

```python
# 6번
netflix[netflix['date_added'].between('2019-05-01', '2019-06-01')].sort_values(by=['date_added'])
```

```python
# 7번
netflix.dropna(subset=['director'])
```
