```python
# 1번
citi_bike = pd.read_csv('citibike.csv',
                        parse_dates=['start_time', 'stop_time'])
citi_bike.head()
```

```python
# 2번
citi_bike['Day of Week'] = citi_bike['start_time'].dt.day_name()
citi_bike.value_counts(subset=['Day of Week']).sort_values(ascending=False)
```

```python
citi_bike['day'] = citi_bike['start_time'].dt.day
citi_bike
```

```python
googan = [0, 7, 14, 21, 28, 30]
citi_bike['googan'] = pd.cut(citi_bike['day'], bins=googan, labels=['1st Week', '2nd Week', '3rd Week', '4th Week', '5th Week'])
citi_bike
```

```python
# 3번
citi_bike['googan'].value_counts().sort_index()
```

```python
# 4번
citi_bike['duration'] = citi_bike['stop_time']-citi_bike['start_time']
citi_bike.head()
```

```python
# 5번
citi_bike['duration'].mean()
```

```python
# 6번
citi_bike['duration'].nlargest(n=5)
```
