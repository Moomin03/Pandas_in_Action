```python
# 그리고, 변수에 값을 넣어줄게요.
employees = pd.read_csv(filepath_or_buffer='employees.csv',
                        parse_dates=['Start Date'])
```

```python
# 정보를 봐볼까요?
employees.info()
```

```python
# Mgmt의 값이 True/False이니, bool 형태로 만들어줄게요.
employees.Mgmt = employees.Mgmt.astype(bool)
```

```python
employees.info()
```

```python
# 그리고 Salary 값에 결측치 값들을 0으로 만들어주도록 하죠.
employees.Salary.fillna(0).head()
```

```python
employees.Salary = employees.Salary.fillna(0).astype(int)
```

```python
employees.info()
```

```python
# 각 열마다의 중복되지 않은 값들의 개수를 보니, Gender와 팀은 카테고리로 만들어도 괜찮겠군요.
employees.nunique()
```

```python
# 먼저 Gender를 카테고리로 만듭시다.
employees['Gender'] = employees['Gender'].astype('category')
```

```python
employees.info()
```

```python
# 그리고, 10개의 팀을 카테고리로 만들죠.
employees['Team'] = employees['Team'].astype('category')
```

```python
employees.info()
```

```python
# First Name이 Maria라는 사람을 찾는 방법입니다.
employees[employees['First Name']=='Maria']
```

```python
# Team이 Finance이 아닌것을 찾는 방법입니다.
employees[employees['Team']!='Finance'].head()
```

```python
# 부정은 ~를 사용할 수 있습니다.
employees[~(employees['Team']=='Finance')].head()
```

```python
# employeess['Mgmt']는 그 값 자체로 참이므로, 해당 값이 True인 데이터만 반환합니다.
employees[employees['Mgmt']].head()
```

```python
# 봉급이 10만 달러를 넘는 사람만 추출합니다.
high_earners = employees[employees['Salary'] > 100000]
high_earners.head()
```

```python
# 여성이며, 팀이 Business Dev인 사람들을 추출할 수 있습니다.
employees[(employees['Gender']=='Female')&(employees['Team']=='Business Dev')].head()
```

```python
# 봉급이 4만달러 이하이고, 일을 시작한 년도가 2015년 1월 1일 이상인 사람들만 추출합니다.
employees[(employees['Salary']<=40000)|(employees['Start Date']>='2015-01-01')].head()
```

```python
# 봉급이 10만달러 이하인 사람들만 추출합니다.
employees[employees['Salary']<100000].head()
```

```python
# 봉급이 10만달러보다 작지 않은 사람들을 추출합니다.
employees[~(employees['Salary']<100000)].head()
```

```python
# 아래 함수들은 연산자를 사용하는 대신, 함수를 사용하여 값을 도출할 수 있습니다.
# employees[employees['Salary']==61933]
employees[employees['Salary'].eq(61933)]
# employees[employees['Salary']!=61933
employees[employees['Salary'].ne(61933)]
# employees[employees['Salary']<100000]
employees[employees['Salary'].lt(100000)]
# employees[employees['Salary']<=100000]
employees[employees['Salary'].le(100000)]
# employees[employees['Salary']>100000]
employees[employees['Salary'].gt(100000)]
# employees[employess['Salary']>=100000]
employees[employees['Salary'].ge(100000)]
```

```python
# 여러 조건을 붙여, or 연산자로 사용할 수 있습니다.
sales = employees['Team']=='Sales'
legal = employees['Team']=='Legal'
mktg = employees['Team']=='Marketing'
employees[sales|legal|mktg]
```

```python
# or 대신에 isin 함수를 사용해서 값을 처리할 수 있습니다.
all_start_teams = ['Sales', 'Legal', 'Marketing']
employees[employees['Team'].isin(all_start_teams)]
```

```python
# between 함수를 사용하면, 각각의 조건에 연산자를 사용하지 않고 사용할 수 있습니다.
employees[employees['Salary'].between(80000, 90000)]
```

```python
# left는 시작, right는 종료일자입니다.
employees[employees['Start Date'].between(
    left='1980-01-01',
    right='1990-01-01'
)]
```

```python
# isnull은 해당 값이 결측치일때 True를 반환합니다.
employees.isnull().head()
```

```python
# notnull은 해당 값이 결측치가 아닐때 True를 반환합니다.
employees.notnull().head()
```

```python
# 전체 열중에서, 결측치의 합을 구합니다.
employees.isnull().sum()
```

```python
employees = pd.read_csv(filepath_or_buffer='employees.csv',
                        parse_dates=['Start Date'])
```

```python
# dropna는 각 행에 어떤 열의 값이라도 결측치가 있으면 해당 행을 버립니다.
employees.dropna()
```

```python
# how='all'의 경우 모든 열의 값이 결측치여야 해당 행을 버립니다.
employees.dropna(how='all')
```

```python
# how='any'의 결과는 일반 dropna의 결과와 같습니다.
employees.dropna(how='any')
```

```python
# subset을 설정하여, 해당 열에 결측치가 있을 때, 행을 버립니다.
employees.dropna(subset=['Gender'])
```

```python
# subset으로 여러개의 열을 선택할 수 있습니다.
employees.dropna(subset=['Start Date', 'Salary'])
```

```python
# thresh는 임계값으로 결측치의 허용범위가 4까지라는 의미입니다.
employees.dropna(thresh=4)
```

```python
# duplicated는 중복인지를 반환합니다.
employees['Team'].duplicated().head()
```

```python
# first는 처음부터 시작해서, 처음 등장하는 값은 False, 한번 이상 등장했던 값은 True로 반한합니다.
employees['Team'].duplicated(keep='first').head()
```

```python
# last의 경우 끝에서 부터 시작합니다.
employees['Team'].duplicated(keep='last')
```

```python
# 중복된 값을 True로 반환한다고 했으니, 중복되지 않고 처음 등장하는 값을 표현하려면 ~를 쓰면 되겠죠?
employees[~employees['Team'].duplicated()]
```

```python
# drop duplicates는 중복된 열을 제거합니다.
employees.drop_duplicates()
```

```python
# subset을 설정하면, Team을 기준으로 중복된 열을 제거합니다.
employees.drop_duplicates(subset=['Team'])
```

```python
# last의 경우, 끝에서부터 중복된 값을 제거합니다.
employees.drop_duplicates(subset=['Team'], keep='last')
```

```python
# keep을 불리언으로 사용할 수도 있는데, 중복된 값이 있는 모든 행을 제거할 수 있습니다.
employees.drop_duplicates(subset=['First Name'],
                          keep=False)
```

```python
# 첫번째 이름이 Duglas이고, Gender가 남자인 사람을 추출하니, 많은 결과가 나왔습니다.
employees[(employees['First Name']=='Douglas')&(employees['Gender']=='Male')]
```

```python
# 이럴경우, subset으로 설정하면 Douglas이면서, 남자인 데이터를 하나만 두고 모두 제거해버리죠.
employees.drop_duplicates(subset=['First Name', 'Gender'])
```
