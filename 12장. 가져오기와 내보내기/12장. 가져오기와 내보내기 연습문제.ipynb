```python
import pandas as pd
```

```python
# 1. show 열의 딕셔너리에 중첩된 에피소드 데이터를 정규화하세요.
# 2. 각각의 show에 대한 개별 데이터프레임을 만드세요.
# 3. 각각의 데이터프레임을 통합 엑셀 문서로 저장하고 sheet_name은 알아서 하세요.
```

```python
tv_show = pd.read_json('tv_shows.json')
tv_show.head()
```

```python
# 1번
pd.json_normalize(data=tv_show['shows'],
                  meta=['show', 'runtime', 'network'],
                  record_path='episodes')
```

```python
# 2번
tv_data = pd.json_normalize(data=tv_show['shows'],
                  meta=['show', 'runtime', 'network'],
                  record_path='episodes')
f_program = tv_data[tv_data['show']=='The X-Files']
s_program = tv_data[tv_data['show']=='Lost']
t_program = tv_data[tv_data['show']=='Buffy the Vampire Slayer']
```

```python
# 3번
excel_file = pd.ExcelWriter(path='tv_show.xlsx')
f_program.to_excel(excel_writer=excel_file, sheet_name='first_program', index=False)
s_program.to_excel(excel_writer=excel_file, sheet_name='second_program', index=False)
t_program.to_excel(excel_writer=excel_file, sheet_name='third_program', index=False)
excel_file.close()
```
