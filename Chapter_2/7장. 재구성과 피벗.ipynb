```python
import pandas as pd
```

```python
# 먼저 데이터를 불러오죠.
pd.read_csv('sales_by_employee.csv').head()
```

```python
# 날짜 데이터도 지정을 해서, 변수에 저장해줄게요.
sales = pd.read_csv(filepath_or_buffer='sales_by_employee.csv',
            parse_dates=['Date'])
```

```python
# pivot_table을 사용하는데 행은 Date를 기준으로, 값들은 Expenses와 Revenue의 값을 사용해 평균을 구합니다.
sales.pivot_table(index='Date',
                  values=['Expenses', 'Revenue'],
                  aggfunc='mean')
```

```python
# 이번엔 합을 구하는 예제입니다.
sales.pivot_table(index='Date',
                  values=['Expenses', 'Revenue'],
                  aggfunc='sum')
```

```python
# columns를 사용하면, Revenue에 대한 값을 더 세분화 시킬 수 있죠.
sales.pivot_table(index='Date',
                  values='Revenue',
                  aggfunc='sum',
                  columns='Name',
                  fill_value=0)
```

```python
# fill_value를 0으로 하면 결측치의 값을 0으로 설정하는 거구요, margins은 전체 데이터의 합을 구하는 겁니다.
sales.pivot_table(index='Date',
                  values='Revenue',
                  aggfunc='sum',
                  columns='Name',
                  fill_value=0,
                  margins=True)
```

```python
# margins에 이름을 설정할 수 있습니다.
sales.pivot_table(index='Date',
                  values='Revenue',
                  aggfunc='sum',
                  columns='Name',
                  fill_value=0,
                  margins=True,
                  margins_name='Total')
```

```python
# count를 사용해서 구할 수도 있구요.
sales.pivot_table(index='Date',
                  columns='Name',
                  values='Revenue',
                  aggfunc='count',
                  fill_value=0)
```

```python
sales.pivot_table(index='Date',
                  columns='Name',
                  values='Revenue',
                  aggfunc=['sum', 'count'],
                  fill_value=0)
```

```python
# 이렇게 각 값에 어떤 함수를 적용할 것인지 선택할 수도 있습니다.
sales.pivot_table(index='Date',
                  columns='Name',
                  values=['Revenue', 'Expenses'],
                  aggfunc={'Revenue':'min', 'Expenses':'min'},
                  fill_value=0)
```

```python
sales.pivot_table(index=['Name', 'Date'],
                  values=['Revenue'],
                  aggfunc='sum')
```

```python
sales.pivot_table(index=['Date', 'Name'],
                  values='Revenue',
                  aggfunc='sum',
                  fill_value=0)
```

```python
by_name_and_date = sales.pivot_table(index='Name',
                  columns='Date',
                  values='Revenue',
                  fill_value=0,
                  aggfunc='sum')
by_name_and_date.head()
```

```python
# stack함수는 열을 행으로 바꿔버립니다.
by_name_and_date.stack()
```

```python
sales_by_customer = sales.pivot_table(index=['Customer', 'Name'],
                                      values='Revenue',
                                      aggfunc='sum')
sales_by_customer.head()
```

```python
# 반대로 unstack() 함수는 행을 열로 바꿔버리죠.
sales_by_customer.unstack()
```

```python
video_game_sales = pd.read_csv(filepath_or_buffer='video_game_sales.csv')
video_game_sales.head()
```

```python
# 가장 중요한 melt함수인데, 기본 데이터 형은 넓게 펼쳐져있잖아요?
# NA, EU, JP, Other을 하나의 카테고리에 넣을 수 있을 것 같지 않나요?
# 그럴때 사용하는게 melt입니다.
# melt는 id_vars에 인덱스에 지정할 열을,
# value_vars에 요약할 열들을 대입합니다.
video_game_sales.melt(id_vars='Name',
                      value_vars=['NA'])
```

```python
# 여러 값들을 넣어줄 수 있는데, 이름이 variable, value잖아요?
video_game_sales.melt(id_vars='Name',
                      value_vars=['NA', 'EU', 'JP', 'Other'])
```

```python
# var_name과 value_name을 통해서 해당 열의 이름을 바꿔줄 수 있습니다.
video_game_sales_by_region = video_game_sales.melt(id_vars='Name',
                                                   value_vars=['NA', 'EU', 'JP', 'Other'],
                                                   var_name='Region',
                                                   value_name='Sales')
video_game_sales_by_region.head()
```

```python
# 각 영역 별, 판매가도 구할 수 있구요.
video_game_sales_by_region.pivot_table(index='Region',
                                       values='Sales',
                                       aggfunc='sum',
                                       fill_value=0)
```

```python
# 자, 여기서 주의해야하는게 원래는 Wii Sports 데이터가 하나였는데, Region에 값에 따라 4배로 늘어났죠?
video_game_sales_by_region[video_game_sales_by_region['Name']=='Wii Sports']
```

```python
# 이 말인 즉, 기존 네게의 열이 행으로 이동했으니, 당연히 행도 4배가 늘어나는거겠죠.
len(video_game_sales_by_region)==len(video_game_sales)*4
```
